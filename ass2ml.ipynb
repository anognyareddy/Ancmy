Distribution and Frequency Analysis

```python
heart_col = 'Heart Disease'

with_disease = (df[heart_col] == 'Presence').sum()
without_disease = (df[heart_col] == 'Absence').sum()

print("Patients with heart disease:", with_disease)
print("Patients without heart disease:", without_disease)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8, 6))
sns.histplot(df['Age'], bins=10, kde=True)
plt.title('Distribution of Age')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.grid(axis='y', alpha=0.75)
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8, 6))
sns.histplot(df['Cholesterol'], bins=10, kde=True)
plt.title('Distribution of Cholesterol Levels')
plt.xlabel('Cholesterol')
plt.ylabel('Frequency')
plt.grid(axis='y', alpha=0.75)
plt.show()
```

 Relationship and Correlation Analysis

```python
correlation_matrix = df.corr(numeric_only=True)
print(correlation_matrix)
```

```python
# Filter pairs with correlation above 0.5 or below -0.5 (excluding self-correlation)
high_corr = correlation_matrix[(correlation_matrix > 0.5) | (correlation_matrix < -0.5)]
print(high_corr)
```

```python
import matplotlib.pyplot as plt
plt.scatter(df['Age'], df['Thallium'], color='purple', alpha=0.6)
plt.title('Age vs Max Heart Rate')
plt.xlabel('Age')
plt.ylabel('Max Heart Rate (thalach)')
plt.grid(True)
plt.show()
```

```python
import numpy as np

# Calculate correlation between Age and Max HR
age_maxhr_corr = df['Age'].corr(df['Max HR'])
print(f"Correlation between Age and Maximum Heart Rate: {age_maxhr_corr:.2f}")
```

```python
# Calculate mean Max HR for patients with and without Heart Disease
mean_maxhr_by_disease = df.groupby('Heart Disease')['Max HR'].mean()
print("\nMean Maximum Heart Rate by Heart Disease Status:")
print(mean_maxhr_by_disease)
```
